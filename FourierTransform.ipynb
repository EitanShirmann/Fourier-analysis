```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python

t = np.arange(-2,2,0.001)

def cos_wave(f):
    return np.cos(2*np.pi*f*t)
```

For each winding frequency create a list of two-tuples, one is rotating vector's magnitude, and the second one is the angle 

```python
#real coordinate
r_cord = []
cos = cos_wave(3)+cos_wave(5)+cos_wave(10)+cos_wave(2)
 
#the grequency range for winding up the function
min_freq_range = 0.0
max_freq_range = 100.0

#the slider of winding speeds
winding_list = np.arange(min_freq_range, max_freq_range, 0.1)

for wind in winding_list:
    # here we have some winding speed

    
    r_cord.append( [(cos[i],    t[i]*wind*2*np.pi)   for i in range(len(t))] )

   
    #   r_cord is now of shape [100,4000,2] - 
    #   100 different winding speeds,
    #   each has 4000 measure points, 
    #   each measure point function series     -  amplitude  (distance from the origin )
    #   and x * winding_frequency * 2pi        - angle theta (angle to the x axis)
```

Get the projections the the axes, to find x,y coordinates of the points

```python
x_cord , y_cord = [], []

#looping through the hundred different winds 
for wind in range(len(r_cord)):
    


    # we now have an individual winding frequency called wind
    x_cord.append(  [amp*np.cos(theta) for (amp, theta) in r_cord[wind]])
    y_cord.append(  [amp*np.sin(theta) for (amp, theta) in r_cord[wind]])
    

    
```

Calculate the means for each winding speed 

```python
mean_list = []
for l in range(len(r_cord)):
    mean = np.sum(x_cord[l])
    mean_list.append(mean)
```

Plot the function
